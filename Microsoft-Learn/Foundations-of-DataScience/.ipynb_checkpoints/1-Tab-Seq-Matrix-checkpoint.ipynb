O pacote numpy, abreviado como np em programas, fornece aos programadores de Python funções poderosas e eficientes para criar e manipular matrizes.
Por exemplo, a função diff computa a diferença entre cada par adjacente de elementos em uma matriz. O primeiro elemento do diff é o segundo elemento menos o primeir

```python
np.diff(highs)
```

Referencia completa: https://numpy.org/doc/stable/reference/

```python
# Cada uma dessas funções usa uma matriz como um argumento e retorna apenas um valor.
np.prod-Multiplicar todos os elementos juntos
np.sum-Adicionar todos os elementos juntos
np.all-Testar se todos os elementos são valores verdadeiros (números diferentes de zero são verdadeiros)
np.any-Testar se pelo menos um dos elementos é um valor verdadeiro (números diferentes de zero são verdadeiros)
np.count_nonzero-Contar o número de elementos diferentes de zero


# Cada uma dessas funções usa uma matriz como um argumento e retorna uma matriz de valores.
np.diff-Diferença entre elementos adjacentes
np.round-Arredonde cada número para o inteiro mais próximo (número inteiro)
np.cumprod-Um produto cumulativo: para cada elemento, multiplique todos os elementos incluídos até o momento
np.cumsum-Uma soma cumulativa: para cada elemento, adicione todos os elementos incluídos até o momento
np.exp-Exponenciar cada elemento
np.log-Pegar o logaritmo natural de cada elemento
np.sqrt-Pegar a raiz quadrada de cada elemento
np.sort-Classificar os elementos

# Cada uma dessas funções usa uma matriz de cadeias de caracteres e retorna uma matriz.
np.char.lower-Colocar cada elemento em minúsculas
np.char.upper-Colocar cada elemento em maiúsculas
np.char.strip-Remover espaços no início ou no final de cada elemento
np.char.isalpha-Se cada elemento é ou não apenas letras (sem números nem símbolos)
np.char.isnumeric-Se cada elemento é ou não apenas numérico (sem letras)

# Cada uma dessas funções usa uma matriz de cadeias de caracteres e uma cadeia de caracteres de pesquisa ; 
# cada uma delas retorna uma matriz.
np.char.count	Contar o número de vezes que uma cadeia de caracteres de pesquisa aparece entre os elementos de uma matriz
np.char.find	A posição dentro de cada elemento em que uma cadeia de caracteres de pesquisa é encontrada primeiro
np.char.rfind	A posição dentro de cada elemento em que uma cadeia de caracteres de pesquisa é encontrada por último
np.char.startswith	Se cada elemento começa ou não com a cadeia de caracteres de pesquisa
```

### Intervalos

Um intervalo é uma matriz de números em ordem crescente ou decrescente, cada um separado por um intervalo regular. Os intervalos são úteis em uma gama surpreendentemente grande de situações.

```python
# Os intervalos são definidos usando a função np.arange, que usa um, dois ou três argumentos: 
# start (início), end (fim) e 'step' (passo)
# np.arange(end): An array starting with 0 of increasing consecutive integers, stopping before end.
np.arange(5)
```

```python
# np.arange(start, end): An array of consecutive increasing integers from start, stopping before end.
np.arange(3, 9)
```

```python
# np.arange(start, end, step): A range with a difference of step between each pair of consecutive values, starting from start and stopping 
# before end.
np.arange(3, 30, 5)
```

```python
np.arange(1.5, -2, -0.5)
```

### # Exemplo

```python
# Exemplo: Fórmula de Leibniz para π (https://en.wikipedia.org/wiki/Gottfried_Wilhelm_Leibniz)
by_four_to_20 = np.arange(1, 20, 4)
by_four_to_20
```

```python
# Para obter uma aproximação precisa de π, usaremos a matriz positive_term_denominators, muito mais longa
positive_term_denominators = np.arange(1, 10000, 4)
positive_term_denominators
```

```python
positive_terms = 1 / positive_term_denominators
negative_terms = 1 / (positive_term_denominators + 2)

# Result
4 * ( sum(positive_terms) - sum(negative_terms) )
```

Aritmética aplicada a um elemento por vez, em pares de matrizes numéricas

O número π é importante em muitas áreas diferentes da matemática. Séculos antes que os computadores fossem inventados, os matemáticos trabalharam para encontrar maneiras simples de aproximar o valor numérico de π. Já vimos a fórmula de Leibniz para 
π. Cerca de meio século antes de Leibniz, o matemático inglês John Wallis (1616–1703) também expressou π em termos de frações simples, como um produto infinito.

```python
#1
even = np.arange(2, 1000001, 2)
one_below_even = even - 1
one_above_even = even + 1

#2
2 * np.prod(even/one_below_even) * np.prod(even/one_above_even)
```
