# Plotar os clássicos

Neste JNB, exploraremos as estatísticas de dois romances clássicos: 
- As Aventuras de Huckleberry Finn de Mark Twain 
- Mulherzinhas de Louisa May Alcott.


Usando o Python, podemos carregar o texto desses livros diretamente da Web.

```python
import requests

# Read two books, fast!

# Livro1
huck_finn_url = 'http://www.gutenberg.org/files/76/76-0.txt'
huck_finn_text = requests.get(huck_finn_url)
huck_finn_chapters = huck_finn_text.split('CHAPTER ')[44:] #lista

# Livro2
little_women_url = 'http://www.gutenberg.org/cache/epub/514/pg514.txt'
little_women_text = requests.get(little_women_url)
little_women_chapters = little_women_text.split('CHAPTER ')[1:] #lista
```

```python
# Colocar a lista de capitulos em uma tabela para ver como cada um começa:

Table().with_column('Chapters', huck_finn_chapters)
```

## Personagens literários

```python
# Visualizar quantas vezes personagens foram mencionados em qualquer ponto do livro
# Count how many times the names Jim, Tom, and Huck appear in each chapter.

counts = Table().with_columns([
        'Jim', np.char.count(huck_finn_chapters, 'Jim'),
        'Tom', np.char.count(huck_finn_chapters, 'Tom'),
        'Huck', np.char.count(huck_finn_chapters, 'Huck')
    ])

# Plot the cumulative counts:
# how many times in Chapter 1, how many times in Chapters 1 and 2, and so on.

cum_counts = counts.cumsum().with_column('Chapter', np.arange(1, 44, 1))
cum_counts.plot(column_for_xticks=3)
plots.title('Cumulative Number of Times Each Name Appears', y=1.08);


# No gráfico, o eixo horizontal mostra os números dos capítulos, e o eixo vertical, 
# quantas vezes cada personagem foi mencionado até determinado capítulo.
```

```python
# The chapters of Little Women, in a table

Table().with_column('Chapters', little_women_chapters)
```

```python
# Counts of names in the chapters of Little Women

counts = Table().with_columns([
        'Amy', np.char.count(little_women_chapters, 'Amy'),
        'Beth', np.char.count(little_women_chapters, 'Beth'),
        'Jo', np.char.count(little_women_chapters, 'Jo'),
        'Meg', np.char.count(little_women_chapters, 'Meg'),
        'Laurie', np.char.count(little_women_chapters, 'Laurie'),

    ])

# Plot the cumulative counts.

cum_counts = counts.cumsum().with_column('Chapter', np.arange(1, 48, 1))
cum_counts.plot(column_for_xticks=5)
plots.title('Cumulative Number of Times Each Name Appears', y=1.08);
```

```python
# Outro tipo de caractere

# In each chapter, count the number of all characters;
# call this the "length" of the chapter.
# Also count the number of periods.

chars_periods_huck_finn = Table().with_columns([
        'Huck Finn Chapter Length', [len(s) for s in huck_finn_chapters],
        'Number of Periods', np.char.count(huck_finn_chapters, '.')
    ])
chars_periods_little_women = Table().with_columns([
        'Little Women Chapter Length', [len(s) for s in little_women_chapters],
        'Number of Periods', np.char.count(little_women_chapters, '.')
    ])
```

```python
# No gráfico abaixo, há um ponto para cada capítulo em cada livro.
# Pontos azuis correspondem ao livro As Aventuras de Huckleberry Finn e os pontos dourados, ao livro Mulherzinhas.
# O eixo horizontal representa o número de pontos, e o eixo vertical, o número de caracteres.

plots.figure(figsize=(6, 6))
plots.scatter(chars_periods_huck_finn.column(1),
              chars_periods_huck_finn.column(0),
              color='darkblue')
plots.scatter(chars_periods_little_women.column(1),
              chars_periods_little_women.column(0),
              color='gold')
plots.xlabel('Number of periods in chapter')
plots.ylabel('Number of characters in chapter');
```

Conclusão

Como tínhamos observado ao examinar os números, o gráfico mostra que muitos capítulos de Mulherzinhas são maiores que os de As Aventuras de Huckleberry Finn, mas alguns não. Contudo, isso também nos mostra algo mais. Observe como os pontos azuis estão praticamente agrupados em torno de uma linha reta, assim como os pontos amarelos. Além disso, parece como se as duas cores de pontos pudessem ser agrupadas em torno da mesma linha reta. 

Agora examine todos os capítulos que contêm cerca de 100 períodos. O gráfico mostra que esses capítulos contêm cerca de 10.000 a 15.000 caracteres aproximadamente. Isso representa cerca de 100 a 150 caracteres por período.

Na verdade, ao examinar o gráfico, parece que, em média, os dois livros tendem a ter aproximadamente entre 100 e 150 caracteres entre os períodos, como estimativa aproximada. Talvez esses dois grandes romances do século XIX estivessem sinalizando algo que é muito familiar para nós hoje: o limite de 140 caracteres do Twitter.
