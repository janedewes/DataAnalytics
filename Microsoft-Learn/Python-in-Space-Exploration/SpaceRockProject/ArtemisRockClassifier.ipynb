# Exploração de dados sobre rochas lunares

```python
print("Artemis Rover Rock Scanner Starting")
```

```python
# Definir variaveis como 0, já que ainda nenhuma rocha foi contada 
# Criar uma lista chamada rockList que armazenará os nomes de cada rocha que o robô encontrou

basalt = 0
breccia = 0
highland = 0
regolith = 0
rockList = []
```

### 1. Ler os dados de rocha lunar de um arquivo de texto para exploração

```python
# Carregar o arq .txt
strPath = "rocks.txt"
fileObject = open(strPath)
line = fileObject.readline()
print(line)
```

```python
# Colocar todas as linhas na var rockList com a fç readlines(), imprimir e fechar:

rockList = fileObject.readlines()

for rock in rockList:
    print(rock)

fileObject.close()
```

### 2. Criar uma função para contar os tipos de rochas lunares encontrados

```python
def countMoonRocks(rockToID):
    global basalt
    global breccia
    global highland
    global regolith

    rockToID = rockToID.lower()

    if("basalt" in rockToID):
        print("Found a basalt\n")
        basalt += 1
    elif("breccia" in rockToID):
        print("Found a breccia\n")
        breccia += 1
    elif("highland" in rockToID):
        print("Found a highland\n")
        highland += 1
    elif("regolith" in rockToID):
        print("Found a regolith\n")
        regolith += 1

    return
```

### 3. Contar o número de rochas lunares chamando uma função Python

```python
for rock in rockList:
    countMoonRocks(rock)
```

### 4. Analisar e resumir dados em Python

```python
# TODO Add a print statement for the other types of rocks: breccia, highland and regolith
print("Number of Basalt: ", basalt)
print("Number of Breccia: ", breccia)
print("Number of Highland: ", highland)
print("Number of regolith: ", regolith)
```

### 5. Os tipos de rocha mais e menos comuns usaremos as funções max() e min()

```python
print("Number of Basalt: ", basalt)
print("Number of Breccia: ", breccia)
print("Number of Highland: ", highland)
print("Number of Regolith: ", regolith)

print("The max number of one type of rock found was:", max(basalt, breccia, highland,regolith))
print("The minimum number of one type of rock found was:", min(basalt, breccia, highland, regolith))
```
