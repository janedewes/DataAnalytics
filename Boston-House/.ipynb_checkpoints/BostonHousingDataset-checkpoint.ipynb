# Exploring the Boston Housing Dataset

Dataset: http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html or
https://archive.ics.uci.edu/ml/machine-learning-databases/housing/
        
Load and return the boston house-prices dataset (regression).

```python
# Imports
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
# Boston dataset is available on scikit-learn, we just need to carry it.
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# Type of boston variable
type(boston)
```

```python
# Dataset shape, 506 Instances(rows) and 13 Attributes(columns)
boston.data.shape
```

```python
# Dataset description
print(boston.DESCR)
```

```python
# Attribute names
print(boston.feature_names)
```

```python
# Transforming the dataset into a pandas DataFrame 
df = pd.DataFrame(boston.data)
df.head()
```

```python
# Transf the column names
df.columns = boston.feature_names
df.head()
```

```python
# boston.target is an array with house prices 
boston.target
```

```python
# Add the house prices on DataFrame (new column with the array above)
df['PRICE'] = boston.target
df.head()
```

```python
# With the new shape
df.shape
```

# Predicting House Prices in Boston

### Y - dependent variable (Boston house prices)

### X - independent or explanatory variables (all other characteristics of the house)

```python
# Importing the Linear Regression module
from sklearn.linear_model import LinearRegression
```

```python
# We don't want the house price as a dependent variable (so we drop the PRICE column)
X = df.drop('PRICE', axis = 1)
X.head()
```

```python
# And define Y (House Price)
Y = df.PRICE
```

```python
# Scatterplot
# Column Rm (df.RM = average number of rooms per house), Y House Price
plt.scatter(df.RM, Y)
plt.xlabel("Average Number of Rooms per House")
plt.ylabel("House Price")
plt.title("Relationship between Number of Rooms and Price")
plt.show()
```

### Machine Learning Model 

```python
# Creating the Linear Regression object
regr = LinearRegression()
```

```python
# Object type
type(regr)
```

```python
# Model training
regr.fit(X, Y)
```

```python
# Coefficient
print("Coefficient: ", regr.intercept_)
print("Coefficient numbers: ", len(regr.coef_))
```

### Prediction 

```python
# Predicting the house price
regr.predict(X)
```

```python
# Comparing original prices vs. predicted prices
plt.scatter(df.PRICE, regr.predict(X))
plt.xlabel("Original Price")
plt.ylabel("Predicted Price")
plt.title("Original Prices vs. Predicted Price")
plt.show()
```

We can see that there are some errors in the prediction of house prices!

```python
MSE?
```

```python
# Calculate the MSE (Mean Squared Error)
mse1 = np.mean((df.PRICE - regr.predict(X)) ** 2)
print(mse1)
```

```python
# Applying linear regression for just one variable and calculating the MSE
regr = LinearRegression()
regr.fit(X[['PTRATIO']], df.PRICE)
mse2 = np.mean((df.PRICE - regr.predict(X[['PTRATIO']])) ** 2)
print(mse2)
```
