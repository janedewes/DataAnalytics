# Regressão e Correlação 

### Duas técnicas relacionadas, com objetivo de estimar o grau de relação existente entre duas variáveis. 

### Regressão: uma equação matemática que descreve o relacionamento entre variáveis 
 - Pode ser linear, polinomial, logística, etc. 
 
### Correlação: o grau de relacionamento entre duas variáveis
- Varia de -1 a 1

```python
import numpy as np
from sklearn import datasets, linear_model 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
production = np.array([877.252, 1125.296, 1317.822, 2420.78, 2625.742, 2686.193, 2916.534, 32189.187, 3368.42, 3375.228, 3518.441, 3552.19, 3617.581, 3618.559])
consumption = np.array([604.685, 682.931, 957.309, 1625.642, 2446.6, 1745.406, 2445.433, 2898.25, 1875.934, 2157.475, 2394.205, 2060.597, 2096.973, 2419.206])

plt.scatter(production, consumption, color ='black')

plt.show()


# OBS: refazer os dados!
```

```python
production = np.array([877.252, 1125.296, 1317.822, 2420.78, 2625.742, 2686.193, 2916.534, 32189.187, 3368.42, 3375.228, 3518.441, 3552.19, 3617.581, 3618.559])
consumption = np.array([604.685, 682.931, 957.309, 1625.642, 2446.6, 1745.406, 2445.433, 2898.25, 1875.934, 2157.475, 2394.205, 2060.597, 2096.973, 2419.206])

# Obj para calcular a regressao linear
regr = linear_model.LinearRegression()
regr.fit(production.reshape(-1, 1), consumption) # faza RL
y = regr.predict(production.reshape(-1, 1))

# y = ax + b
print('Coeficientes: \n', regr.coef_) # valor da eq que acompanha x
print('Intercept: \n',regr.intercept_) # valor de b

# Plots outputs
plt.scatter(production, consumption, color='black')
plt.plot(production, y, color='blue', linewidth=3) # linha

plt.show()
```
